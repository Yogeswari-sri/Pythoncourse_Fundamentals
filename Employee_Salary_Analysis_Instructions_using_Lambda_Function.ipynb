Case Study – Employee Salary Analysis
Instructions

Create a list called salary Store employee salaries inside the list

Use Built-in Functions to:

Find maximum salary

Find minimum salary

Find total salary

Find number of employees using len()

Use lambda with map() to: Increase each salary by 10%

Use lambda with filter() to: Display salaries greater than 50000

Display:

Original salary list

Updated salary list

High salary list

Maximum salary

Minimum salary

Total salary

```python
salary=[52000,45000,30000,60000,15000]

# Built-in Functions
max(salary)
min(salary)
sum(salary)
len(salary)

#Lambda Function with map()
Updated_salary=list(map(lambda x:x + x*1.10 , salary))

#Lambda Function with filter()
Highest_salary=list(filter(lambda x: x > 50000 , Updated_salary))

#Display_Outputs:
print(f"Original salary list {salary}")
print(f"Updated salary list {Updated_salary}")
print(f"Highest_salary list {Highest_salary}")
print(f"Maximum salary {max(salary)}")
print(f"Minimun salary {min(salary)}")
print(f"Total salary {sum(salary)}")
print(f"Number of Employees {len(salary)}")
```
