**-- LIBRARY FINE--**

```python
student_name=input("Enter student name:")
book_name=input("Enter the Book name:")
days_late=int(input("Enter how many days later on:"))
fine_per_day=float(input("Enter per day fine amount:"))
Total_fine=days_late*fine_per_day
print(f"LIBRARY FINE DETAILS \nStudentname: {student_name} \nBooksname: {book_name} \nDayslate: {days_late} \nPerdayfineamount: {fine_per_day} \nTotalfine: Rs {Total_fine}")

#type casting
print(type(student_name))
print(type(book_name))
print(type(days_late))
print(type(fine_per_day))
```
