# **Case Study – Student Marks Analysis using Function and Loop**

**Instructions**

Create a list called marks

Create a function called marks_analysis() and pass the marks list to the function

Inside the function:

Use a for loop to display all marks

Find total marks

Find average marks


Return total and average marks, call the function, and display: Student Marks, Total Marks,  Average Marks

```python
# Marks List
marks =[100,95,90]

# Function Creation
def marks_analysis(marks):

    # for loop
    for i in marks:
        print(i)

    # Operations
    total = sum(marks)
    averagemarks= total / len(marks)

    # Return Values
    return total,averagemarks

# Function Call
total,averagemarks = marks_analysis(marks)

# Output
print(f"TotalMarks : {total}")
print(f"AverageMarks : {averagemarks}")
```
