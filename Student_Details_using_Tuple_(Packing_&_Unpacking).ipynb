**Student Details using Tuple (Packing & Unpacking)**

Instructions:

Create a tuple named student and store the following details: name, age, course

Unpack the tuple into separate variables: name, age, course

Display each value separately using print().



```python
# Packing and Unpacking Tuples

# Packing (multiple values into a single tuple)
student = ("Yoga", 32, "Data Analytics")

print("Packed Tuple:", student)

# Unpacking (extracting values from tuple into variables)
name, age, course = student

print("Name:", name)
print("Age:", age)
print("Course:", course)
```
