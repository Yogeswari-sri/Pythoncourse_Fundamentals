# **Case Study – String Analysis using List Comprehension**

**Instructions **

Create a string variable called text

Store any sentence or word inside the variable

Use list comprehension to:

Convert all characters into uppercase

Extract only vowels from the string

Extract only consonants from the string


Display:

Original string

Uppercase characters

Vowels list

Consonants list

String without spaces

Print the output in a clean format using print() statements

```python
# String Analysis using List Comprehension

text = "data analytics"
text1= "Hello world123"

# Convert to uppercase
upper_text = [x.upper() for x in text]

# Extract vowels
vowels = [x for x in text if x in "aeiou"]

# Extract consonants
consonants = [x for x in text if x.isalpha() and x not in "aeiou"]

# Remove spaces
remove_spaces = [x for x in text if x != " "]

# Extract Digits
remove_digits =[t1.isalnum() for t1 in text1]

print("Uppercase Letters :", upper_text)
print("Vowels            :", vowels)
print("Consonants        :", consonants)
print("Without Spaces    :", remove_spaces)
print("Without Digits    :",remove_digits)
```
