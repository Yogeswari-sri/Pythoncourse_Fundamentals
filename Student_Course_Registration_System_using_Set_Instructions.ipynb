Case Study – Student Course Registration System using Set
Instructions:

Create a set named courses with course names: Python, SQL, Power BI, Python, Excel

Perform set operations: • Add course using add()
• Add multiple courses using update()
• Remove course using remove() and discard()

Display all courses and total number of courses using len()



```python
course={"python","SQL","Power BI","Python","Excel"}

#Add course using add()
course.add("IT")
print("After added",course)

#Add multiple courses using update()
course.update(["ML","Medicine"])
print("After updates",course)

#Remove course using remove() and discard()
#course.remove() - if not data values are there error comes so safely used discard() to remove a single values

course.discard("ML")
print("After discard",course)

len(course)
print("Length of courses are:",course)
```
