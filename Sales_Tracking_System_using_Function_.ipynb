# **Case Study: Sales Tracking System**

A retail store records sales transactions every day. The manager wants a Python function that can:

Accept any number of sales amounts.

Calculate the current day's total sales.

Maintain a running total of all sales made till date.

Display both the current sale total and the cumulative sales total.

Task:

Write a Python function total_sale(*sales) that uses a global variable to keep track of the total sales till date.

```python
cum_total=26000
def total_sale(*sales):
  current_total = sum(sales)
  global cum_total
  cum_total += current_total

  return current_total,cum_total


current_total,cum_total=total_sale(3000,200,1000,5000)


print(f"Current_total {current_total}")
print(f"Cummulative_total {cum_total}")

```
