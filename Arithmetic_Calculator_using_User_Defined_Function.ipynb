# **Case Study – Arithmetic Calculator using User Defined Function**

**Instructions**

Create a function called arth_cal()

Pass two parameters to the function

Inside the function:

Find addition

Find subtraction

Find multiplication

Return all three values using multiple return statements


Take two numbers from the user using input()

Store returned values in separate variables

Display:

Sum

Difference

Multiplication

```python
def arth_cal(a,b):
  add = a+b
  sub = a-b
  mul = a*b
  return add,sub,mul
arth_cal(10,2)
print(f"Sum : {arth_cal(10,2)}")
print(f"Difference : {arth_cal(30,15)}")
print(f"Multiplication : {arth_cal(30,2)}")
```
