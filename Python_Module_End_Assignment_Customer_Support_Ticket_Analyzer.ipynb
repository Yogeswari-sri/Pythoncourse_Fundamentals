Step 3: Text Cleaning for Issue Descriptions


❖ Clean all issue descriptions by applying:

➢ Remove punctuation (.,!?-)

➢ Convert multiple spaces → single space

➢ Remove leading/trailing spaces

➢ Convert text to lowercase

➢ Replace slang/shorthand (example: "ok" → "okay")

```python
import string
def clean_issue(text):
    text = text.translate(str.maketrans('', '', string.punctuation))  # remove punctuation
    text = " ".join(text.split())  # normalize spaces
    text = text.strip().lower()    # trim + lowercase
    text = text.replace("ok", "okay")  # slang replace
    return text

ticket_data["Issue_Description"] = [clean_issue(desc) for desc in ticket_data["Issue_Description"]]
ticket_data
```

Step 4: Keyword-Based Issue Insights

Create a function:

def count_tickets_with_word(word):

```python
def count_tickets_with_word(word):
    word = word.lower()
    count = 0
    for data in ticket_data["Issue_Description"]:
        if word in data.lower():
            count += 1
    return count

# Usage
print("Total poor:", count_tickets_with_word("poor"))
print("Total good:", count_tickets_with_word("good"))
print("Total slow:", count_tickets_with_word("slow"))
print("Total excellent:", count_tickets_with_word("excellent"))
```

```python
# 2. Priority Analysis
def count_tickets_with_priority(word):
    word = word.lower()
    count = 0
    for data in ticket_data["Priority"]:
        if word in data.lower():
            count += 1
    return count


print("Total count of High priority:", count_tickets_with_priority("High"))
print("Total count of medium priority:", count_tickets_with_priority("Medium"))
print("Total count of low priority:", count_tickets_with_priority("Low"))
```

```python
# 3. Ticket with Longest Issue Description
length=[len(data.split()) for data in ticket_data["Issue_Description"]]
max_ticket=max(length)


print(f"Ticket No: {ticket_data['Ticket_No'][max_ticket]}")
print(f"Customer name: {ticket_data['Customer_Name'][max_ticket]}")
print(f"Issue description: {ticket_data['Issue_Description'][max_ticket]}")
print(f"Word Count: {length[max_ticket]}")
```

```python
#4,unique words
unique_words = set()
for data in ticket_data["Issue_Description"]:
    unique_words.update(data.split())


print(f"Count of Unique Words: {len(unique_words)}")
print(f"Word List: {sorted(unique_words)}")
```
