Case Study: Student Performance Analysis Using NumPy

A school wants to analyze the marks of students in 5 subjects using NumPy.

You need to:

Store student marks using NumPy arrays

Calculate:

Total marks, Average marks, Highest marks, Lowest marks,Percentage

Find:

Students who scored above average, Students who failed, Subject-wise average

```python
# STUDENT PERFORMANCE ANALYSIS USING NUMPY

# Import NumPy library
import numpy as np

marks = np.array([
    [100, 89, 76, 89,97],   # Student A
    [78, 85, 90, 88, 76],   # Student B
    [65, 70, 60, 72, 68],   # Student C
    [90, 92, 88, 95, 91],   # Student D
    [45, 50, 55, 48, 52],   # Student E

])

students=np.array(["A","B","C","D","E"])
print(students)


print("--------- STUDENT MARKS --------\n")
print(marks)

total_marks=np.sum(marks,axis=1)
print("Total marks are : ",total_marks)

Average = np.mean(marks,axis=1)
print("Average  : ",Average)

max_marks=np.max(marks,axis=1)
print("Maximum marks ", max_marks)

min_marks=np.min(marks,axis=1)
print("Minimum marks ", min_marks)

percentage=(total_marks/500)*100
percentage

overall_average = np.mean(total_marks)
print("Overall_Average " ,overall_average)

student_above_average=students[total_marks > overall_average]
print("Student above average marks ",student_above_average)

failed_students=students[np.any(marks < 50,axis=1)]
print("failed students  below 50 : ",failed_students)

subject_average=np.mean(marks,axis=0)
print("Subject average :",subject_average)

# Display Student-wise Report

print("\n========== STUDENT REPORT ==========")

for i in range(len(students)):

    print("\nStudent Name :", students[i])

    print("Marks :", marks[i])

    print("Total Marks :", total_marks[i])

    print("Average Marks :", Average[i])

    print("Highest Mark :", max_marks[i])

    print("Lowest Mark :", min_marks[i])

    print("Percentage :", percentage[i])


# Rank Students Based on Total Marks
# argsort() sorts in ascending order
# [::-1] converts to descending order

ranking = np.argsort(total_marks)[::-1]     #index wise sorting

print("\n========== STUDENT RANKING ==========")

for rank, index in enumerate(ranking, start=1):

    print(f"Rank {rank} : Student {students[index]}) "
    f"-> Total Marks = {total_marks[index]}")



```
