Create a dictionary named `student` with the following keys:
   name, age, course

2. Store multiple students' data using lists as values.

3. Add one new student details to the dictionary:

   • Append new name to name list  

   • Append new age to age list  

   • Append new course to course list  

4. Display all student details using dictionary keys.

5. Access and display:

   • First student's details using indexing

   • Newly added student's details using indexing

```python

students = {
    "name": ["yoga","sri","Haashraa"],
    "age": [32, 5, 13],
    "course": ["IT", "AI", "Medicalscience"]
}


# Add one new student(new name)
students["name"].append("kadar")
print(f"new names are {students['name']}")
students["age"].append(30)
students["course"].append("IT")

# Display all data
print("names are ", students["name"])
print("ages:", students["age"])
print("courses are", students["course"])

# First student details
print("\n----- First Student -----")
print(f"first stud name:", students["name"][0])
print(f"Age:", students["age"][0])
print(f"Course:", students["course"][0])


# Second student details
print("\n----- Second Student -----")
print("Name:", students["name"][1])
print("Age:", students["age"][1])
print("Course:", students["course"][1])

# Newly added student details
print("\n----- New Student -----")
print("Name:", students["name"][3])
print("Age:", students["age"][3])
print("Course:", students["course"][3])
```
