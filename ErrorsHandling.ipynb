# **Basic Python Errors**

| Error Type            | When It Occurs                   | Example                    |
| --------------------- | -------------------------------- | -------------------------- |
| **SyntaxError**       | Invalid Python syntax            | `if True`                  |
| **IndentationError**  | Missing or incorrect indentation | `if True:\nprint("Hello")` |
| **NameError**         | Variable not defined             | `print(age)`               |
| **TypeError**         | Using incompatible data types    | `10 + "20"`                |
| **ValueError**        | Invalid value for a function     | `int("abc")`               |
| **ZeroDivisionError** | Dividing by zero                 | `10 / 0`                   |
| **IndexError**        | List index out of range          | `[10,20,30][5]`            |
| **KeyError**          | Dictionary key not found         | `{"name":"John"}["age"]`   |
| **AttributeError**    | Method doesn't exist for object  | `100.append(10)`           |
| **FileNotFoundError** | File does not exist              | `open("data.txt")`         |

**try and except in Python**

try and except are used for exception handling.

 They help prevent a program from crashing when an error occurs.

try:

    # Code that may cause an error
except:

    # Code that runs if an error occurs

```python
#Handling Division by Zero
n = int(input("Enter the number to divide from 10 : "))
try:
    num = 10 / n
    print(num)
except:
    print("Cannot divide by zero")
```

```python
#Handling Specific Error
try:
    num = int(input("Enter a number: "))
    result = 100 / num
    print(result)

except ZeroDivisionError:
    print("Cannot divide by zero")

except ValueError:
    print("Please enter a valid number")
```

```python
try:
    num = int(input("Enter a number: "))
    result = 100 / num

except ValueError:
    print("Please enter a valid number")

except ZeroDivisionError:
    print("Cannot divide by zero")

else:
    print("Result:", result)

finally:
    print("Program Ended")
```
