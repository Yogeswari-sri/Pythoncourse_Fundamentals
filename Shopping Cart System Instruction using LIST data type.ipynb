Case Study – Shopping Cart System
Instructions:

Create a list named cart with initial items: Milk, Bread, Eggs, Apple

Perform list operations:

• Add multiple items using extend()

• Insert item at index 2 using insert()

• Remove item Apple using remove()

• Remove last item using pop()

Use list methods:

• index() to find position of Eggs

• count() to count occurrences of Milk

Apply indexing:

• Display first item

• Display last item

Apply slicing:

• Display first 3 items from the list

Find total number of items using len()

Display all outputs using print()

```python
cart=["Milk","Bread","Eggs","Apple"]
print(f"Cart List: {cart}")

cart.extend(["Orange","wheat"])
print(f"After extend cart:{cart}")

cart.insert(2,"Cheese")
print(f"After inserting 2nd place Items:{cart}")

cart.remove('Apple')
print(f"Remove Apple Items:{cart}")

cart.pop()
print(f"Last Items removed{cart}")

index_Eggs=cart.index("Eggs")
print(f"position of eggs:{index_Eggs}")

count_milk=cart.count("Milk")
print(f"Counting occurence of Milk:{count_milk}")

cart[:1]
print(f"First Item:{cart[:1]}")

cart[-1:]
print(f"Last Item:{cart[-1:]}")

cart[:3]
print(f"Fistt 3 Items displayed:{cart[:3]}")

length=len(cart)
print(f"Total number of items:{length}")
```
