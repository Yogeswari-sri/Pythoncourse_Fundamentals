**Instructions:**

1. Take a string input from the user using input().

2. Convert the text into lowercase using lower().

3. Use a for loop to iterate through each character in the string.

4. Count:
   • Total vowels  
   • Total consonants  
   • Total digits  
   • Total special characters  

5. Use:
   • isalpha() for alphabets  
   • isdigit() for numbers  

6. Display all counts using print()

```python
Text = input("Enter your string : ").lower()
vowels = consonant = digits = special_character =0

for t in Text:
  if t in "aeiou":
    vowels +=1
  elif t.isalpha():
    consonant += 1
  elif t.isdigit():
    digits += 1
  else:
    special_character += 1
print("Total vowels",vowels)
print("Total consonants",consonant)
print("Total digits",digits)
print("Total special character",special_character)

```
