```python
# Email Username Generator
full_name=input('Enter the full_name')
domain_name=input("Enter domain_name")

lower=full_name.lower()


email_name=lower.replace(" ", ".")

email=email_name + "@" + domain_name + ".com"

print('email',(email))

```
