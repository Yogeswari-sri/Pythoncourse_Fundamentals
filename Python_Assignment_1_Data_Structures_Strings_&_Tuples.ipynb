```python
Text1="Hello"
Name = input("Enter Your name:")

str_con=Text1 + " " +Name
print(f"{str_con}")

# ** Now, concatenate string 3, "Welcome to Python programming,” to the existing string and display the output string.**

str3="Welcome to Python programming"

str3_con=str_con+ "," +str3
print(f"{str3_con}")
# a. Print the first character of the string.
str3_con[0]
print(f"first character of the string: {str3_con[0]}")

# b. Print the last character of the string.
str3_con[-1]
print(f"The last character of the string: {str3_con[-1]}")

# c. Print the first 5 characters of the string.
str3_con[:5]
print(f"The first 5 characters of the string: {str3_con[:5]}")

# d. Print the last 11 characters of the string.
str3_con[-11:]
print(f"The last 11 character of the string: {str3_con[-11:]}")

# e. Print the string in reverse.
str3_con[::-1]
print(f"Reverse string: {str3_con[::-1]}")

# f. Use slicing and print the word “Python” from the existing string.
str3_con[22:28]
print(f"{str3_con[22:28]}")

strM="Python beginner tutorial"
print(f"{strM}")

# a. Convert the sentence to uppercase.
strM.upper()
print(f"uppercase sentence:{strM.upper()}")

# b. Convert the sentence to lowercase.
strM.lower()

print(f"lowercase sentence : {strM.lower()}")

# c. Use Capitalize and return the sentence to the original input form.
strM.capitalize()

print(f" Capitalize sentence: {strM.capitalize()}")

# d. Count the total number of occurrences of character ‘t’ in the string.

strM.count("t")

print(f"Total occurances of t character : {strM.count("t")}")

 #e. Replace all occurrences of “Python” with “Data Analytics” in the input string

strM="Python beginner tutorial."
strM.replace("Python","Data Analytics")

print(f"Replaced String : {strM.replace("Python","Data Analytics")}")

tuple_1=(10,20,30)
tuple_2=(40,50,60)

print(f"Tuple_1 values :{tuple_1}")
print(f"Tuple_2 values :{tuple_2}")


# a. Concatenate the two tuples and store it in “t_combine”

t_combine=tuple_1+tuple_2
print(f"Combined Tuple Values : {t_combine}")

# b. Repeat the elements of “t_combine” 3 times

t_combine*3
print(f"Repeat 3 Times combining elements: {t_combine*3}")

# c. Access the 3rd element from “t_combine”

t_combine[2]
print(f"Third element: {t_combine[2]}")

# d. Access the first three elements from “t_combine”

t_combine[:3]
print(f"First 3 elements : {t_combine[:3]}")

# e. Access the last three elements from “t_combine”

t_combine[-3:]

print(f"Last 3 elements : {t_combine[-3:]}")
```
