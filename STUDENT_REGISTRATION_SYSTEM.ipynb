**---- STUDENT REGISTRATION SYSTEM----**

```python
Student_name=input("Enter the name")
Student_age=int(input("Enter the age"))
Course_name=input("Enter the Coursename")
Course_fee=float(input("Enter the fees"))
gst=0.18*Course_fee
Total_fee=Course_fee+gst
print(f"STUDENT DETAILS \nStudentname: {Student_name} \nStudentage: {Student_age} \nCoursename: {Course_name} \nCoursefee: Rs {Course_fee} \nGST: {gst} \nTotalfees: Rs {Total_fee}")

# TYPE CASTING
print(type(Student_age))
print(type(Course_fee))
print(type(Course_name))
print(type(Student_name))
```
