**Online Order Summary using Tuple(packing and Unpacking) **



Instructions:

Create a tuple named order with the following details: order_id, product_name, quantity, price, status

Unpack the tuple into separate variables: order_id, product, quantity, price, status

Display all order details using print() and f-string

```python
# PACKING ORDER USING TUPLES
order_tuple=(101,'Smart watch',5,8000,'delivered')
print(f'Your ordes is:',order_tuple)

# UNPACKING ORDER TUPLE
order_id,product_name,quantity,price,status=order_tuple

print(f"Your orders are:{order_id}")
print(f"Name of the product:{product_name}")
print(f"Their quantiy:{quantity}")
print(f"product price:{price}")
print(f"status are:{status}")

# AFTER UPDATING AGAIN PACKING USING TUPLES
price=8500

order_tuple=(order_id,product_name,quantity,price,status)
print(f"\n-----ORDER DETAILS-----")
print(f"Your orders are:{order_tuple}")
print(type(order_tuple))
```
