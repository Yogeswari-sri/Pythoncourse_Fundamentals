##  **5. Case Study: Bank Loan Eligibility System**
**Instructions:**

1. Create variables:
   customer_name, age, income, credit_score, existing_loan

2. Use conditional statements (if, elif, else) to:
   • Approve loan if age ≥ 21, income ≥ 30000, credit_score ≥ 700, and no existing loan  
   • Conditionally approve loan if credit_score ≥ 600 and income ≥ 40000  
   • Otherwise reject the loan application  

3. Display customer details and loan status using print()

```python
Customer_Name = input("Enter Customer name: ")
Age = int(input("Enter Your age: "))
Income = int(input("Enter Your Income: "))
Credit_score = int(input("Enter Your credit score values: "))
Existing_loan = input("Enter any previous loan status (Yes/No): ").lower()

if Age >= 21 and Income >= 40000:
    if Credit_score >= 800 and Existing_loan == 'no':
        status = "Approved"
    elif Credit_score >= 700 and Income >= 45000:
        status = "Conditionally Approved"
    else:
        status = "Not Approved Your Loan"
else:
    status = "Not Eligible"

print("-------Loan Eligibility Details---------")
print(f"Customer_Name : {Customer_Name}")
print(f"Age : {Age}")
print(f"Income per Month : {Income}")
print(f"Your Credit_score values : {Credit_score}")
print(f"Your Existing loan status : {Existing_loan}")
print(f"Loan Approved Status : {status}")

```
