Task 1: While Loop & Control Statements -Number Guessing Game
Problem Statement: Create a Python program that implements a simple number guessing
game using a while loop. The program should make use of control statements such as else,
break, and continue.
Instructions:
1. Set Up the Game:
★ Generate a random number between 1 and 10 that the user has to guess.
Import random and use randint function.
import random
# Generate random number between 1 and 10
secret_number = random.randint(1, 10)

```python
import random
# Generate random number between 1 and 10
secret_number = random.randint(1, 10)

guessing_number = attempts = 0


while attempts < 3:
    guessing_number = int(input("Guess the Number (between 1 and 10)"))
    if guessing_number < 1 or guessing_number > 10:
        print("If the guess is out of Valid Range")
        continue

    if secret_number > guessing_number:
      print("Guessing_number is lower than Secret_Number")
    elif secret_number < guessing_number:
      print("Guessing_number is greater than Secret_Number")
    else:
      print("cONGRATULATIONS ! You Guessed The correct number !")
      break

    attempts += 1
if attempts == 3 and guessing_number != secret_number:
  print("Better luck next time!,user runs out of attempts without guessing the correct number")
```

Task 3: Function- BMI Calculator
Problem Statement:


Create a Python program that calculates the Body Mass Index (BMI).
Hint: BMI = weight (kg) / [height (m)]2

```python
def calculate_bmi():
  weight = float(input("Enter your weight in kg : "))
  height = float(input("Enter your height in meters : "))
  BMI = weight  / (height ** 2)
  return round(BMI,2)

calculate_bmi()
```
